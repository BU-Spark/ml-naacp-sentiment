CLEANING TEXT

```python
def process_text(text):
    text = text.lower().translate(str.maketrans('', '', string.punctuation))
    
    tokens = word_tokenize(text)
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in tokens]
    return lemmatized_words
#print (process_text("4 Big guys and they bust on my eyes, they eat my ass jsut like apple pies."))
```

```python
df['text'].astype(str)  # Convert to string type if not already
df['text'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df['text'].str.lower()
df['text'].str.strip()
lemmatizer= WordNetLemmatizer()
df['lemmatized_text'] = df['text'].apply(lambda row: ' '.join(nltk.word_tokenize(row)))
df['lemmatized_text'] = df['lemmatized_text'].astype(str)
data_type = df['lemmatized_text'].dtype
print("Data type of lemmatized_text:", data_type)
first_object = df['lemmatized_text'].iloc[0]

# Use the dir() function to list attributes and methods of the object
object_attributes = dir(first_object)
print("Attributes of the object:", object_attributes)

df.head()
```
